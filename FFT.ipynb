```python
samp_rate = 1000.0
samp_time = 1.0
samples = samp_rate*samp_time
t = np.linspace(0, samp_time, samples)
f1, f2 = 53, 60
y = np.sin((2*np.pi)*f1*t) + np.sin((2*np.pi)*f2*t) + \
    np.random.normal(scale = 3.5, size=len(t)) # Add some noise
plot(t, y)
xlim(0,0.2)
```

```python
fft = np.fft.rfft(y)
x_freq = np.fft.fftfreq(2*len(fft), d=1/samp_rate)[:len(fft)]
plot(x_freq, np.abs(fft)**2)
xlim(0,100)
xlabel('Frequency (Hz)', fontsize=18)
```
