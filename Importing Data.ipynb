## Importing data

Many experiments record data in text or binary files.  In this tutorial, we'll go through importing data so that it can be analyzed.

```python
%pylab inline
```

The first thing to do is figure out the format of the data.  Open up the example data file, `test.txt`.  You should notice that there is a bunch of information at the beginning.  This is called metadata, data about your data.  We see the trigger counts, output counts, sample rate, etc.  We'll ignore these for now, but we'll come back and see how to get the metadata.

Below that, you'll see rows and columns of data.  We'll start with the easiest method of loading data.  From the data file, we see that the first five rows should be skipped since the data starts at the sixth row.  Also, each column is delimited by tabs, so we set the delimiter to `\t` which is the escape character for tabs.

```python
data = np.loadtxt('test.txt', delimiter='\t', skiprows=5)
data[0,:]
```

You can see that the first row of `data` is the first data row in `test.txt`.  Maybe we only want the time of the first pulse and the time of the second pulse.  We can import only those two columns and name them by setting the `dtype`.  Now we can read off the code clearly, we plot the difference in the time of the pulses.

```python
data = np.loadtxt('test.txt', delimiter='\t', skiprows=5, usecols=(0,3))
data.dtype = [('time1','f8'), ('time2','f8')]
_ = hist(data['time2'] - data['time1'], bins=100)
```

```python
dt = data['time2'] - data['time1']
counts, x, _ = hist(dt, bins=200, range = (0,5e-6))
x = x[:-1]*1e6 # Convert to microseconds
bg = np.mean(counts[125:])
```

```python
plot(x, counts-bg)
```

```python
plot(x, log(counts-bg))
```

```python
muons = counts - bg
roi = np.where((1<x) * (x<2.3))
(m, b), cov = np.polyfit(x[roi], np.log(muons)[roi], 1, cov=True)
sig_m, sig_b = np.sqrt(cov.diagonal())
plot(x, log(muons))
plot(x[roi], m*x[roi] + b)
print("Muon lifetime = {:1.3f} microseconds with uncertainty {:1.3f} microseconds".
      format(m, sig_m))
```
